```python
import qrcode
import image
qr = qrcode.QRCode(
    version = 15,
    box_size = 10,
    border = 5
)

data ="https://www.linkedin.com/in/akchaurasiya-092/"
qr.add_data(data)
qr.make(fit = True)
img = qr.make_image(fill = "black", back_color = "gray")
img.save("ash.png")
```
